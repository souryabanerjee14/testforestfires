### Algerian Forest Fire Dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset= pd.read_csv('Algerian_forest_fires_dataset.csv')
```

```python
dataset.head()
```

```python
dataset.info()
```

## Data Cleaning


```python
## missing values
dataset[dataset.isnull().any(axis=1)]
```

```python
dataset = dataset.drop(122)
```

```python
dataset.reset_index(drop=True,inplace=True)
```

```python
dataset[dataset.isnull().any(axis=1)]
```

The dataset is converted into two sets based on Region from 123rd index, we can make a new coloumn based on the region

1: Bejaia Region Dataset

2: "Sidi Bel Abbes Region Dataset"

Add new column with region

```python
dataset.loc[:122,"Region"]=0
dataset.loc[122:,"Region"]=1
df =dataset
```

```python
df.info()
```

```python
df[['Region']]=df[['Region']].astype(int)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
##Removing NullValues

df = df.dropna().reset_index(drop=True)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.iloc[[122]]
```

```python
df = df.drop(122).reset_index(drop=True)
```

```python
df.iloc[[122]]
```

```python
df.columns
```

```python
### fix spaces in column name
df.columns= df.columns.str.strip()
df.columns
```

```python
df.info()
```

### Change the required colums as integer data type

```python
df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']]=df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']].astype(int)
```

```python
df.info()
```

### Changing the other columns to float data datatype

```python
objects=[features for features in df.columns if df[features].dtypes=='O']
```

```python
for i in objects:
    if i !='Classes':
        df[i]=df[i].astype(float)
```

```python
df.info()
```

```python
df.describe()
```

```python
df.head()
```

```python
## Lets save the cleaned dataset
df.to_csv('Algerian_cleaned.csv', index=False)
```

### Exploratory Data Analysis


```python
## drop day, month and year
df_copy= df.drop(['day','month','year'], axis =1)
```

```python
df_copy.head()
```

```python
df_copy['Classes'].unique()
```

```python
## Encoding the categories in classes
df_copy['Classes'] = np.where(df_copy['Classes'].astype(str).str.contains('not fire'), 0, 1)
```

```python
df_copy.head()
```

```python
## categories in classes
df_copy['Classes'].value_counts()
```

```python
## Plot density plot for all features

df_copy.hist(bins=50,figsize=(20,15))
plt.show()
```

```python
## Percentage for Pie Chart
percentages= df_copy['Classes'].value_counts(normalize=True)*100
```

```python
##Plotting piechart
classlabels=["Fire", "Not Fire"]
plt.figure(figsize=(12,7))
plt.pie(percentages, labels=classlabels, autopct='%1.1f%%' )
plt.title('Pie Chart of Classes')
plt.show()
```

```python
## Monthly Fire Analysis

dftemp = df[df["Region"]==0].copy()
dftemp['Classes'] = np.where(dftemp['Classes'].astype(str).str.contains('not fire'), 'not fire', 'fire')
plt.subplots(figsize=(13,6))
sns.set_style('whitegrid')
sns.countplot(x='month',hue='Classes',data=dftemp)
plt.title("Bejiah")
```

```python
dftemp = df[df["Region"]==1].copy()
dftemp['Classes'] = np.where(dftemp['Classes'].astype(str).str.contains('not fire'), 'not fire', 'fire')
plt.subplots(figsize=(13,6))
sns.set_style('whitegrid')
sns.countplot(x='month',hue='Classes',data=dftemp)
plt.title("Sidi bel")
```
